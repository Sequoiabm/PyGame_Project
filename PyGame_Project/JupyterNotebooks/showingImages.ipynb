```python
# Images variables

# Title screen
titleImage = "MainTitle.png"

# Level one images
image1 = "Lvl1_1.png"
image2 = "Lvl1_2.png"
image3 = "Lvl1_3.png"
image4 = "Lvl1_4.png"
```

```python
# Function to display images:
import matplotlib.pyplot as plt
import os

def showImage(image):
    os.system('clear')
        # Load the image
    image = plt.imread(image)

    # Display the image
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
print("               Welcome to Macro: The Ultimate Adventure")
showImage(titleImage)
choice = input("Enter 'y' to start the game")
choice

if choice == 'y':
    print("You are at work tasked with delivering a package, would you like to wait or hit the road?")
    showImage(image1)
    choice1 = input("[1]Wait\n[2]Hit the road")
    choice1

    if choice1 == '1':
        print("You decide to wait, nothing much here.")
        showImage(image2)
    elif choice1 == '2':
        print("You hit the road, look at your map.")
        showImage(image4)
```
